```python
import pandas as pd
dados_teste = pd.read_json('dados/dados_teste.json')
df_plano = pd.json_normalize(pd.read_json('dados/dados_teste.json', encoding='utf-8')['data'])
df_plano["valorTotalEstimado"].sum()
```

```python
valor_total = df_plano["valorTotalEstimado"].sum()
print(f"💰 O valor total estimado de todas as licitações é: R$ {valor_total:,.2f}")
```

```python
df_plano["dataInclusao"] = pd.to_datetime(df_plano["dataInclusao"])
df_plano["dataPublicacaoPncp"] = pd.to_datetime(df_plano["dataPublicacaoPncp"])
df_plano.info()
```
